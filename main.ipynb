```python
# importing important libraries needed for the program to run
import csv
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import car_process as cp
import Car_TUI as TUI

def load_data(file_path):   
    # specify file-path
    file_path = "Car_manufacture/CarPrice.csv"
    TUI.started(f"Reading data from {file_path}")
    # create an empty list with variable named "details". This is the list where the csv file will be read into.
    details = []
    # Display message
    print("Program started......")
    print("Loading data.........\n", end="")
    with open(file_path) as csv_file:
            csv_reader = csv.reader(csv_file)
            headings = next(csv_reader)
   # using for loop to access each line from the file  
            for line in csv_reader:
               details.append(line)
    TUI.completed()
    return(details)

   
        
def run():
    # Creating variable for the file name
    details = load_data ("Car_manufacture/CarPrice.csv")
    num_details = len(details)
    print(f"{len(details)} records successfully loaded.")
    data = np.loadtxt("Car_manufacture/CarPrice.csv",delimiter=",", dtype=str)
    df = pd.read_csv("Car_manufacture/CarPrice.csv",header=0) #load into a data frame
    
    while True:
        selection = TUI.display_menu()
        print(f"You have selected option: {selection}\n")
        if selection.upper() == "I":
            cp.retrieve_car_details_by_id(details)
        elif selection.upper() == "II":
            cp.get_cars_with_specified_cylinder_number(details)
        elif selection.upper() == "III":
            cp.retrieve_cars_for_a_specified_car_body(details)
        elif selection.upper()  == "IV":
            cp.retrieve_specific_columns(details)
        elif selection.upper() == "V":
            cp.retrieve_car_names_alphabetically(details)
        elif selection.upper() == "VI":
            cp.retrieve_sales_summary(details)
        elif selection.upper() == "VII":
            cp.retrieve_top_5_car_sale(details)
        elif selection.upper()  == "VIII":
            cp.retrieve_selected_car_details(details)
        elif selection.upper()  == "IX":
            cp.display_no_of_cars_per_fuelsystem(details)
        elif selection.upper()  == "X":
            cp.display_horsepower_of_top_5_car_sale(details)
        elif selection.upper() == "XI":
            cp.display_customer_selection(details)
        elif selection  == "exit":
            break
        else:
            print("Error!!! Invalid option")

if __name__ == "__main__":
    run()   
```
